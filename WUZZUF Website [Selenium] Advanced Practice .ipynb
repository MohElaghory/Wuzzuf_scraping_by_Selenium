```python
# Another application is [wuzzuf website], if we need to search on ['python', 'java', 'webscraping'] jobs:

# Chrome driver Path
chromepath= r"chromedriver.exe"

option= webdriver.ChromeOptions()      # open google chrome session

option.add_argument('-disk-cache-size= 3000000000')   # OPTIONAL,    --->  make memory cache size = 3000000000                                                                                                                                                                                                                                                                                                                         
driver= webdriver.Chrome(executable_path= chromepath, options= option)   # open driver


positions_list= ['python', 'java', 'data science']

for i in positions_list:
    driver.get('https://www.google.com/')          # Open Google website
    element= driver.find_element_by_css_selector("input[class= 'gLFyf gsfi']")      # بتاع جوجل search أنا هنا بقف علي ال
    element.send_keys('WUZZUF')                    # ودي اللي هسرش عليها WUZZUF أنا هنا بعد مابقف علي السيرش بكتب جواه كلمه
    element.submit()             # بقي عشان ادخل عليها Enter عايز ادوس WUZZUF بعد مابكتب كلمه

    WUZZUF_url= driver.find_element_by_css_selector("div[class= 'yuRUbf']").find_element_by_css_selector('a').get_attribute('href')         # Here i will get the [first link] after searched at WUZZUF
    driver.get(WUZZUF_url)          # Enter to this [first link]
    
    driver.find_element_by_css_selector("input[class= 'search-bar-input']").send_keys(i)           # [python, java, data science] سواء بقي i بكتب في السيرش الوظيفه اللي هي wuzzuf أنا هنا بعد مابدخل علي موقع 
    driver.find_element_by_css_selector("button[class= 'btn btn-primary search-btn']").click()     # عشان يجبلي الوظايف اللي بسرش عليها search بعد ما بكتب الوظيفه اللي بسرش عنها هنا بقي بدوس علي زرار
    
    # scrape job_titles:
    job_titles= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")            # job_titles منهم واللي هما بالنسبالي دول ال text وده لان التاج ده فيه منه أكتر من واحد بنفس الكلاس ده انا عندي 15 واحد وبعد كده بلوب عليهم وبطلع ال find_element مش find_elements بكلاس معين فعشان كده بنعمل h2 والوظيفه الواحده التاج بتاعها اسمه job_title أنا عندي في الصفحه الواحده 15 وظيفه يعني 15 
    job_titles_lst= [element.text for element in job_titles]
    
    # scrape company_names:
    company_names= driver.find_elements_by_css_selector("a[class= 'css-17s97q8']")         # find_element مش find_elements بكلاس معين وبما ان التاج ده بنفس كلاسه في منه 15 زيه فبالتالي بعمل a فأنا هنا بشوف الوظايف موجوده في انهي تاج لقيته في التاج اللي اسمه job_titles نفس الكلام زي ال
    company_names_lst= [element.text.replace('-', '').strip() for element in company_names]   # عشان اشيل المسافات الزياده اللي علي يمين وشمال اسم الشركه بحيث يرجعلي الاسم بس بدون مسافات وزيادات strip وهنا انا بعمل تنضيف لاسم الشركه لان الاسم بيرجعلي في اخره شرطه - فبشلها بأني بستبدلها ب مسافه وبعد كده بعمل company_name واللي هيبقي بالنسبالي ده ال text بعد كده هلوب علي تاج تاج منهم وهطلع منه ال
    
    # scrape job_location:
    job_location= driver.find_elements_by_css_selector("span[class= 'css-5wys0k']")        # نفس الكلام زي فوق
    job_location_lst= [element.text for element in job_location]
    
    # scrape job_posted:
    job_posted_new= driver.find_elements_by_css_selector("div[class= 'css-4c4ojb']")       # نفس الكلام زي فوق
    job_posted_new_lst= [element.text for element in job_posted_new]
    t.sleep(3)
    job_posted_old= driver.find_elements_by_css_selector("div[class= 'css-do6t5g']")       # نفس الكلام زي فوق
    job_posted_old_lst= [element.text for element in job_posted_old]
    t.sleep(3)
    job_posted_all= [*job_posted_new_lst, *job_posted_old_lst]           # فبكده هيبقوا ليست واحده job_posted_all داخل الليست job_posted_old_lst و job_posted_new_lst واللي معناها فضي محتويات الليست  unbacking ال* يعني
    
    # scrape job_type:
    job_type= driver.find_elements_by_css_selector("div[class= 'css-1lh32fc']")            # نفس الكلام زي فوق
    job_type_lst= [element.text.replace('\n', ' - ') for element in job_type]              # ' - ' ب '\n' هنضف بأني هستبدل ال
    
    # scrape skills:
    job_skills= []
    containers= driver.find_elements_by_css_selector("div[class= 'css-1gatmva e1v1l3u10']")       # كلها skills الكبير اللي شايل ال division ده ال
    for container in containers:       # بتاعت الوظيفه skills المهم هضم العناصر دي كلها مع بعض لانهم كلهم بيرجعوا ال skills اللي بيشيل بردو بعض ال a ثم العنصر span ده فيه اكتر من div ولكن التاني مش الاول الان ال span ثم العنصر Experiances واللي بيشيل ال a ده علي العناصر اللي فيه زي division هلوب علي ال
        skills= container.find_element_by_css_selector("div[class= 'css-y4udm8']").find_element_by_css_selector("div:nth-child(2)").text
        skills= skills.replace('\n', '')        # اللي هي علامه النزول لسطر جديد \n من ال text بنضف ال
        job_skills.append(skills)
    
    # scrape job_link
    containers= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")     # Job_Skills نفس الكلام زي ال
    links= []
    for container in containers:
        job_link= container.find_element_by_css_selector("a[class= 'css-o171kl']").get_attribute('href')
        links.append(job_link)
        
    
    # Save all data in csv files:
    from itertools import zip_longest
    from csv import writer

    file_list= [job_titles_lst, company_names_lst, job_location_lst, job_posted_all, job_type_lst, job_skills, links]
    exported= zip_longest(*file_list)

    with open(f'{i} jobs_first_page.csv', 'w', encoding="utf-8") as csv_file:
        wr= writer(csv_file)
        wr.writerow(['Job Title', 'Company Name', 'Job Location', 'Job Posted', 'Jop Type', 'Job Skill', 'Job Link'])
        wr.writerows(exported)
        
```

```python
# Chrome driver Path
chromepath= r"chromedriver.exe"

option= webdriver.ChromeOptions()      # open google chrome session

option.add_argument('-disk-cache-size= 3000000000')   # OPTIONAL,    --->  make memory cache size = 3000000000
driver= webdriver.Chrome(executable_path= chromepath, options= option)   # open driver

positions_list= ['python', 'java', 'data science']

for i in positions_list:
    
    driver.get('https://www.google.com/')          # Open Google website
    element= driver.find_element_by_css_selector("input[class= 'gLFyf gsfi']")      # بتاع جوجل search أنا هنا بقف علي ال
    element.send_keys('WUZZUF')                    # ودي اللي هسرش عليها WUZZUF أنا هنا بعد مابقف علي السيرش بكتب جواه كلمه
    element.submit()             # بقي عشان ادخل عليها Enter عايز ادوس WUZZUF بعد مابكتب كلمه

    WUZZUF_url= driver.find_element_by_css_selector("div[class= 'yuRUbf']").find_element_by_css_selector('a').get_attribute('href')         # Here i will get the [first link] after searched at WUZZUF                                                                                                                                                                                                                                                                                                                                                                                     
    driver.get(WUZZUF_url)          # Enter to this [first link]

    driver.find_element_by_css_selector("input[class= 'search-bar-input']").send_keys(i)            # [python, java, data science] سواء بقي i بكتب في السيرش الوظيفه اللي هي wuzzuf أنا هنا بعد مابدخل علي موقع
    driver.find_element_by_css_selector("button[class= 'btn btn-primary search-btn']").click()      # عشان يجبلي الوظايف اللي بسرش عليها search بعد ما بكتب الوظيفه اللي بسرش عنها هنا بقي بدوس علي زرار
    driver.find_element_by_css_selector("button[class= 'css-18rqtbx ezfki8j0']").click()            # واللينك ده أنا محتاجه عشان هلوب عليه تحت عشان يلوبلي علي الصفحات كلها url أنا هنا بدوس علي الصفحه الاولي اللي انا فيها عشان يجبلي اللينك كامل فوق في ال


    current_url= driver.current_url                # لأني هحتاج المتغير ده تحت في اللوب اللي بتلف علي الصفح كلها current_url أسمه variable أنا هنا بجيب لينك الصفحه اللي انا واقف فيه دلوقتي واللي هيبقي أول صفحه في صفحات الوظايف وبحفظه في 

    page_limit= driver.find_element_by_css_selector("li[class= 'css-8neukt']").text[-3:]     # text[-4:] ده أخر 3 حروف او ارقام اللي هم يعني عدد الوظايف وطبعا انا بقول 3 لأن اكيد الوظايف مش هتعدي 999 وظيفه ولكن مش شرط ممكن يكونوا اكتر من 1000 يعني 4 ارقام فساعتها هعمل text بيضم الصفجه بيبقي فيها كام وظيفه وفي اخره بيبقي مكتوب عد الوظايف اللي في كل الصفحات فأحنا هنا بنبقي عايزين من ال text دي بتبقي عباره عن li بكلاس معين ف ال li ده بيبقي موجود في اسفل الصفحه خالص في تاج اسمه jobs المتاحة للوظيفه اللي بسرش عليها فدايماً عدد ال jobs هنا أنا بجيب عدد كل ال
    nums_of_pages= int(page_limit) // 15 + 2              # طبعاً انا بعمل كل ده عشان انا اجيب عدد الصفح الوظايف عشان الوب عليها صفحه صفحه وناخد منها الوظايف فعشان كده انا هنا بقسم علي ال 15 لان الصفحه الواحده فيها 15 وظيف وفلما بقسم عدد الوظايف علي عدد وظاسف الصفحه الواحده اللي هم 15 فبالتالي هيطلعلي عدد الصفحات كام صفحه وفي الاخر هلوب عليهم وطبعاً انا هنا بقسم ب // عشان يطلعلي الناتج الصحيح بدون رقم عشري ولكن ان هعتبر الرقم العشري ده كأنه صفحه فعشان كده بضيف اتنين يعني عملت +2

    Job_Title= []
    Company_Names= []
    Job_Location= []
    Job_Postet= []
    Job_Type= []
    Job_Skills= []
    Job_Links= []

    for pg in range(1, nums_of_pages):
        job_titles= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")      # job_titles منهم واللي هما بالنسبالي دول ال text وده لان التاج ده فيه منه أكتر من واحد بنفس الكلاس ده انا عندي 15 واحد وبعد كده بلوب عليهم وبطلع ال find_element مش find_elements بكلاس معين فعشان كده بنعمل h2 والوظيفه الواحده التاج بتاعها اسمه job_title أنا عندي في الصفحه الواحده 15 وظيفه يعني 15 
        job_titles_lst= [element.text for element in job_titles]
        Job_Title.extend(job_titles_lst)          # Job_Title بتاعتنا اللي بره اللي اسمها list في ال extend فهاخدها وهعملها job_titles_lst اللي اسمها list كلهم يرجعوا في ال job_titles بعد ما 

        # scrape company_names:
        company_names= driver.find_elements_by_css_selector("a[class= 'css-17s97q8']")   # find_element مش find_elements بكلاس معين وبما ان التاج ده بنفس كلاسه في منه 15 زيه فبالتالي بعمل a فأنا هنا بشوف الوظايف موجوده في انهي تاج لقيته في التاج اللي اسمه job_titles نفس الكلام زي ال
        company_names_lst= [element.text.replace('-', '').strip() for element in company_names]    # عشان اشيل المسافات الزياده اللي علي يمين وشمال اسم الشركه بحيث يرجعلي الاسم بس بدون مسافات وزيادات strip وهنا انا بعمل تنضيف لاسم الشركه لان الاسم بيرجعلي في اخره شرطه - فبشلها بأني بستبدلها ب مسافه وبعد كده بعمل company_name واللي هيبقي بالنسبالي ده ال text بعد كده هلوب علي تاج تاج منهم وهطلع منه ال
        Company_Names.extend(company_names_lst)   # Company_Names بتاعتنا اللي بره اللي اسمها list في ال extend فهاخدها وهعملها company_names_lst اللي اسمها list كلهم يرجعوا في ال company_names بعد ما 

        # scrape job_location:
        job_location= driver.find_elements_by_css_selector("span[class= 'css-5wys0k']")    # نفس الكلام زي فوق
        job_location_lst= [element.text for element in job_location]
        Job_Location.extend(job_location_lst)

        # scrape job_posted:
        job_posted_new= driver.find_elements_by_css_selector("div[class= 'css-4c4ojb']")   # نفس الكلام زي فوق
        job_posted_new_lst= [element.text for element in job_posted_new]
        t.sleep(3)
        job_posted_old= driver.find_elements_by_css_selector("div[class= 'css-do6t5g']")   # نفس الكلام زي فوق
        job_posted_old_lst= [element.text for element in job_posted_old]
        t.sleep(3)
        job_posted_all= [*job_posted_new_lst, *job_posted_old_lst]       # فبكده هيبقوا ليست واحده job_posted_all داخل الليست job_posted_old_lst و job_posted_new_lst واللي معناها فضي محتويات الليست  unbacking ال* يعني
        Job_Postet.extend(job_posted_all)

        # scrape job_type:
        job_type= driver.find_elements_by_css_selector("div[class= 'css-1lh32fc']")        # نفس الكلام زي فوق
        job_type_lst= [element.text.replace('\n', ' - ') for element in job_type]          # ' - ' ب '\n' هنضف بأني هستبدل ال
        Job_Type.extend(job_type_lst)

        # scrape skills:
        job_skills= []
        containers= driver.find_elements_by_css_selector("div[class= 'css-1gatmva e1v1l3u10']")       # كلها skills الكبير اللي شايل ال division ده ال
        for container in containers:       # بتاعت الوظيفه skills المهم هضم العناصر دي كلها مع بعض لانهم كلهم بيرجعوا ال skills اللي بيشيل بردو بعض ال a ثم العنصر span ده فيه اكتر من div ولكن التاني مش الاول الان ال span ثم العنصر Experiances واللي بيشيل ال a ده علي العناصر اللي فيه زي division هلوب علي ال
            skills= container.find_element_by_css_selector("div[class= 'css-y4udm8']").find_element_by_css_selector("div:nth-child(2)").text
            skills= skills.replace('\n', '')        # اللي هي علامه النزول لسطر جديد \n من ال text بنضف ال
            job_skills.append(skills)
        Job_Skills.extend(job_skills)

        # scrape job_link
        containers= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")     # Job_Skills نفس الكلام زي ال
        links= []
        for container in containers:
            job_link= container.find_element_by_css_selector("a[class= 'css-o171kl']").get_attribute('href')
            links.append(job_link)
        Job_Links.extend(links)

        driver.get(current_url[:-1]+str(pg))      # يدخلي علي الصفحه اللي بعد كده if وطبعاً الامر ده عشان بعد كل حاله
        
        
    # Save all data in csv files:
    from itertools import zip_longest
    from csv import writer

    file_list= [Job_Title, Company_Names, Job_Location, Job_Postet, Job_Type, Job_Skills, Job_Links]
    exported= zip_longest(*file_list)

    with open(f'{i} jobs_all_pages_1.csv', 'w', encoding="utf-8") as csv_file:
        wr= writer(csv_file)
        wr.writerow(['Job Title', 'Company Name', 'Job Location', 'Job Posted', 'Jop Type', 'Job Skill', 'Job Link'])
        wr.writerows(exported)
        
```

```python
# Another application is [wuzzuf website], if we need to search on ['python', 'java', 'webscraping'] jobs, but here we will scraping all jobs from all available pages of jobs:

# Chrome driver Path
chromepath= r"chromedriver.exe"

option= webdriver.ChromeOptions()      # open google chrome session

option.add_argument('-disk-cache-size= 3000000000')   # OPTIONAL,    --->  make memory cache size = 3000000000
driver= webdriver.Chrome(executable_path= chromepath, options= option)   # open driver


positions_list= ['python', 'java', 'data science']

for i in positions_list:
    driver.get('https://www.google.com/')          # Open Google website
    element= driver.find_element_by_css_selector("input[class= 'gLFyf gsfi']")      # بتاع جوجل search أنا هنا بقف علي ال
    element.send_keys('WUZZUF')                    # ودي اللي هسرش عليها WUZZUF أنا هنا بعد مابقف علي السيرش بكتب جواه كلمه
    element.submit()             # بقي عشان ادخل عليها Enter عايز ادوس WUZZUF بعد مابكتب كلمه

    WUZZUF_url= driver.find_element_by_css_selector("div[class= 'yuRUbf']").find_element_by_css_selector('a').get_attribute('href')         # Here i will get the [first link] after searched at WUZZUF                                                                                                                                                                                                                                                                                                                                                                                     
    driver.get(WUZZUF_url)          # Enter to this [first link]
    
    driver.find_element_by_css_selector("input[class= 'search-bar-input']").send_keys(i)            # [python, java, data science] سواء بقي i بكتب في السيرش الوظيفه اللي هي wuzzuf أنا هنا بعد مابدخل علي موقع
    driver.find_element_by_css_selector("button[class= 'btn btn-primary search-btn']").click()      # عشان يجبلي الوظايف اللي بسرش عليها search بعد ما بكتب الوظيفه اللي بسرش عنها هنا بقي بدوس علي زرار
    driver.find_element_by_css_selector("button[class= 'css-18rqtbx ezfki8j0']").click()            # واللينك ده أنا محتاجه عشان هلوب عليه تحت عشان يلوبلي علي الصفحات كلها url أنا هنا بدوس علي الصفحه الاولي اللي انا فيها عشان يجبلي اللينك كامل فوق في ال
    
    # تحت Scraping اللي هعملهم informations فيهم كل ال extend دول فاضيين عشان هأ lists أنا هنا بعمل ال
    Job_Title= []
    Company_Names= []
    Job_Location= []
    Job_Postet= []
    Job_Type= []
    Job_Skills= []
    Job_Links= []
    
    def scraping():
        # scrape job_titles:
        job_titles= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")      # job_titles منهم واللي هما بالنسبالي دول ال text وده لان التاج ده فيه منه أكتر من واحد بنفس الكلاس ده انا عندي 15 واحد وبعد كده بلوب عليهم وبطلع ال find_element مش find_elements بكلاس معين فعشان كده بنعمل h2 والوظيفه الواحده التاج بتاعها اسمه job_title أنا عندي في الصفحه الواحده 15 وظيفه يعني 15 
        job_titles_lst= [element.text for element in job_titles]
        Job_Title.extend(job_titles_lst)          # Job_Title بتاعتنا اللي بره اللي اسمها list في ال extend فهاخدها وهعملها job_titles_lst اللي اسمها list كلهم يرجعوا في ال job_titles بعد ما 

        # scrape company_names:
        company_names= driver.find_elements_by_css_selector("a[class= 'css-17s97q8']")   # find_element مش find_elements بكلاس معين وبما ان التاج ده بنفس كلاسه في منه 15 زيه فبالتالي بعمل a فأنا هنا بشوف الوظايف موجوده في انهي تاج لقيته في التاج اللي اسمه job_titles نفس الكلام زي ال
        company_names_lst= [element.text.replace('-', '').strip() for element in company_names]    # عشان اشيل المسافات الزياده اللي علي يمين وشمال اسم الشركه بحيث يرجعلي الاسم بس بدون مسافات وزيادات strip وهنا انا بعمل تنضيف لاسم الشركه لان الاسم بيرجعلي في اخره شرطه - فبشلها بأني بستبدلها ب مسافه وبعد كده بعمل company_name واللي هيبقي بالنسبالي ده ال text بعد كده هلوب علي تاج تاج منهم وهطلع منه ال
        Company_Names.extend(company_names_lst)   # Company_Names بتاعتنا اللي بره اللي اسمها list في ال extend فهاخدها وهعملها company_names_lst اللي اسمها list كلهم يرجعوا في ال company_names بعد ما 


        # scrape job_location:
        job_location= driver.find_elements_by_css_selector("span[class= 'css-5wys0k']")    # نفس الكلام زي فوق
        job_location_lst= [element.text for element in job_location]
        Job_Location.extend(job_location_lst)

        # scrape job_posted:
        job_posted_new= driver.find_elements_by_css_selector("div[class= 'css-4c4ojb']")   # نفس الكلام زي فوق
        job_posted_new_lst= [element.text for element in job_posted_new]
        t.sleep(3)
        job_posted_old= driver.find_elements_by_css_selector("div[class= 'css-do6t5g']")   # نفس الكلام زي فوق
        job_posted_old_lst= [element.text for element in job_posted_old]
        t.sleep(3)
        job_posted_all= [*job_posted_new_lst, *job_posted_old_lst]       # فبكده هيبقوا ليست واحده job_posted_all داخل الليست job_posted_old_lst و job_posted_new_lst واللي معناها فضي محتويات الليست  unbacking ال* يعني
        Job_Postet.extend(job_posted_all)

        # scrape job_type:
        job_type= driver.find_elements_by_css_selector("div[class= 'css-1lh32fc']")        # نفس الكلام زي فوق
        job_type_lst= [element.text.replace('\n', ' - ') for element in job_type]          # ' - ' ب '\n' هنضف بأني هستبدل ال
        Job_Type.extend(job_type_lst)

        # scrape skills:
        job_skills= []
        containers= driver.find_elements_by_css_selector("div[class= 'css-1gatmva e1v1l3u10']")       # كلها skills الكبير اللي شايل ال division ده ال
        for container in containers:       # بتاعت الوظيفه skills المهم هضم العناصر دي كلها مع بعض لانهم كلهم بيرجعوا ال skills اللي بيشيل بردو بعض ال a ثم العنصر span ده فيه اكتر من div ولكن التاني مش الاول الان ال span ثم العنصر Experiances واللي بيشيل ال a ده علي العناصر اللي فيه زي division هلوب علي ال
            skills= container.find_element_by_css_selector("div[class= 'css-y4udm8']").find_element_by_css_selector("div:nth-child(2)").text
            skills= skills.replace('\n', '')        # اللي هي علامه النزول لسطر جديد \n من ال text بنضف ال
            job_skills.append(skills)
        Job_Skills.extend(job_skills)

        # scrape job_link
        containers= driver.find_elements_by_css_selector("h2[class= 'css-m604qf']")     # Job_Skills نفس الكلام زي ال
        links= []
        for container in containers:
            job_link= container.find_element_by_css_selector("a[class= 'css-o171kl']").get_attribute('href')
            links.append(job_link)
        Job_Links.extend(links)

    page_limit= driver.find_element_by_css_selector("li[class= 'css-8neukt']").text[-3:]     # text[-4:] ده أخر 3 حروف او ارقام اللي هم يعني عدد الوظايف وطبعا انا بقول 3 لأن اكيد الوظايف مش هتعدي 999 وظيفه ولكن مش شرط ممكن يكونوا اكتر من 1000 يعني 4 ارقام فساعتها هعمل text بيضم الصفجه بيبقي فيها كام وظيفه وفي اخره بيبقي مكتوب عد الوظايف اللي في كل الصفحات فأحنا هنا بنبقي عايزين من ال text دي بتبقي عباره عن li بكلاس معين ف ال li ده بيبقي موجود في اسفل الصفحه خالص في تاج اسمه jobs المتاحة للوظيفه اللي بسرش عليها فدايماً عدد ال jobs هنا أنا بجيب عدد كل ال
    nums_of_pages= int(page_limit) // 15 + 2              # طبعاً انا بعمل كل ده عشان انا اجيب عدد الصفح الوظايف عشان الوب عليها صفحه صفحه وناخد منها الوظايف فعشان كده انا هنا بقسم علي ال 15 لان الصفحه الواحده فيها 15 وظيف وفلما بقسم عدد الوظايف علي عدد وظاسف الصفحه الواحده اللي هم 15 فبالتالي هيطلعلي عدد الصفحات كام صفحه وفي الاخر هلوب عليهم وطبعاً انا هنا بقسم ب // عشان يطلعلي الناتج الصحيح بدون رقم عشري ولكن ان هعتبر الرقم العشري ده كأنه صفحه فعشان كده بضيف اتنين يعني عملت +2
    
    current_url= driver.current_url         # لأني هحتاج المتغير ده تحت في اللوب اللي بتلف علي الصفح كلها current_url أسمه variable أنا هنا بجيب لينك الصفحه اللي انا واقف فيه دلوقتي واللي هيبقي أول صفحه في صفحات الوظايف وبحفظه في 
    
    for pg in range(1, nums_of_pages):
        # Next دي بحيث ألوب علي صفحه صفحه ومع كل صفحه اخلصها ادوس if في الصفحه الاولي غير اللي في الصفحه التانيه والتالته غير اللي في باقي الصفح التانيين في الاختلاف مابينهم في اسم الكلاس بتاع كل زرار فعشان كده عملت حالات Next دي لان الزرار if أنا هنا عامل حالات 
        # عشان ينتقل للصفحه اللي بعديها Next لكل معلومات الوظيفه وبعد كده بدوس علي زرار scraping اللي بتعمل function فأنا هنا بقوله لو انا في الصفحه الأولي فبستدعي ال
        if driver.current_url == current_url:     # اللي في اول صفحه ليه كلاس لواحده Next الزرار
            scraping()
            driver.find_element_by_css_selector("button[class= 'css-zye1os ezfki8j0']").click()
            t.sleep(3)      # للموقع والدنيا تقف Unconnect للصفحه اللي بعديها عشان ميحصلش Next بستني 3 ثواني بعد مابدوس 
            
        # اللي في الصفحه التانيه والتالته نفس الكلاس Next الزرار 
        # لو انا في الصفحه التانيه او التالته جمع المعلومات وبعد كده انتقل للصفحه اللي بعديها
        elif driver.current_url == current_url[:-1]+'1' or driver.current_url == current_url[:-1]+'2':
            scraping()
            # الثامن child اللي رقمه الثامن فعشان كده هيبقي ال li والزرار ده جوه التاج اللي اسمه next طبعاً الزرار اللي انا عايز ادوس عليه هو زرار ال
            driver.find_element_by_css_selector("div[class= 'css-if9uys']").find_element_by_css_selector("li:nth-child(8)").find_element_by_css_selector("button[class= 'css-zye1os ezfki8j0']").click()                                                                                                                    
            t.sleep(3)
        
        # اللي في باقي الصفحات شبه بعضيهم ليهم نفس الكلاس Next الزرار
        # لو انا في الصفحات اللي بعد التالتهبدايه من الصفحه الرابعه لحد الاخر جمع المعلومات وبعد كده انتقل للصفحه اللي بعديها
        else:
            scraping()
            # الثامن child اللي رقمه الثامن فعشان كده هيبقي ال li والزرار ده جوه التاج اللي اسمه next طبعاً الزرار اللي انا عايز ادوس عليه هو زرار ال
            driver.find_element_by_css_selector("div[class= 'css-if9uys']").find_element_by_css_selector("li:nth-child(9)").find_element_by_css_selector("button[class= 'css-zye1os ezfki8j0']").click()                                                                                                                    
            t.sleep(3)
            
        driver.get(current_url[:-1]+str(pg))     # يدخلي علي الصفحه اللي بعد كده if وطبعاً الامر ده عشان بعد كل حاله
        t.sleep(3)
    
    
    # Save all data in csv files:
    from itertools import zip_longest
    from csv import writer

    file_list= [Job_Title, Company_Names, Job_Location, Job_Postet, Job_Type, Job_Skills, Job_Links]
    exported= zip_longest(*file_list)

    with open(f'{i} jobs_all_pages_2.csv', 'w', encoding="utf-8") as csv_file:
        wr= writer(csv_file)
        wr.writerow(['Job Title', 'Company Name', 'Job Location', 'Job Posted', 'Jop Type', 'Job Skill', 'Job Link'])
        wr.writerows(exported)
```
